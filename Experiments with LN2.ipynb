```python
import numpy as np
import matplotlib.pyplot as plt

radius = 20e-2
speed_rotation = 115   # degrees per second
tangential_speed = 2*np.pi*radius/360*115

print(f'The tangential speed is {tangential_speed:.3f} m/s.')

# Resolution with SVS no lens
dx = 2.2e-6
N = 10e3
zs = np.linspace(3e-2, 12e-2)
wav = 355e-9

opp = N / 2 * dx
NAs = opp / np.sqrt(opp**2 + zs**2)
rdx = 2.4 * wav / NAs

plt.plot(zs, rdx)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

wavel = 532e-9
pixel = 20e-6
MO = ["Navitar 4k 4X", "Navitar 4k 9X"]
magnification = [4, 9]
Nx = 1024
zs = np.linspace(1e-3, 15e-3, 100)
resolutions = np.zeros((zs.size, len(magnification)))

for k in range(len(MO)):
    eff_pixel = pixel / magnification[k]
    FOV = Nx * eff_pixel
    NA = FOV / 2 / np.sqrt((FOV / 2) ** 2 + zs ** 2)
    resolutions[:, k] = 1.22 * wavel/NA

plt.plot(zs, resolutions)
plt.show()
```
