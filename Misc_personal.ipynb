# Probability of winning freebies in Aliexpress

Binomial distribution

```python
from scipy import stats
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook

# Probability of winning 1 time
trials = np.arange(1,356*2,4)
wins = 1
p = 1/np.array([200, 500, 1000, 2000])
pmfs = [stats.binom.pmf(wins, trials, ap) for ap in p]

fig, ax = plt.subplots()
for knt, pmf in enumerate(pmfs):
    ax.plot(trials, 100*pmf, label='{:.0f}'.format(1/p[knt]))
ax.set_title('Probability of winning 1 time')
ax.set_ylabel('Percentage [%]')
ax.legend(title='No of competitors')
ax.set_xlabel('Trials')
plt.show()

# Probability of winning 1 or more times
wins = 0
pmfs = [1-stats.binom.pmf(wins, trials, ap) for ap in p]

fig, ax = plt.subplots()
for knt, pmf in enumerate(pmfs):
    ax.plot(trials, 100*pmf, label='{:.0f}'.format(1/p[knt]))
ax.set_title('Probability of winning 1 or more times')
ax.legend(title='No of competitors')
ax.set_ylabel('Percentage [%]')
ax.set_xlabel('Trials')
plt.show()
```
