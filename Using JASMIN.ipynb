## Storage

There are 4 types of disk storage:
- SSD : manage small files, non parallel
    - **/home/users** : backed up, persistent, 100GB
    - /work/scratch-nopw : non-persistent (28 days)
- Fast parallel disk (PFS) : not good with small files
    - /group_workspaces/jasmin2
    - /gws/pw/j05
    - /work/scratch-pw
- Scale Out File Storage (SOF) : large volumes of data (not small files), regular use, non-parallel
    - **gws/nopw/j04** : /gws/nopw/j04/holography/users
    - work/xfc/volX
- High Performance Object Storage (HPOS) : cloud storage
