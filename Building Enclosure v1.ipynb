## Structure

10. Use 2 30mm (RS 390-1798) angle bracket sets to place the back top bar.
11. Use 4 30mm (RS 390-1798) angle bracket sets to place the 2 mid-height side bars with length 490 mm.

### Door

### LEDs

- The red LED showing laser pulses goes in the control unit (8 mm holes). 20 mA, 2V.
- The green and yellow LEDs go on the exterior.
- Red LED in exterior to show "laser allowed" with same electric node.
- Use in open-drain output with 3.3V and 102ohm resistance.

#### LEDs modes

1. Initialisation : slow blink on 3 LEDs.
2. Door closed:
    1. green on, yellow and red off.
3. Door open:
    1. NOT alignment mode : internal red and green off, yellow slow blink.
    2. alignment mode : fast blink red and green. Yellow off.
    3. turn on internal red if heater is on or hot.
4. Main heater on or hot: yellow on, overrides others.

Nucleo board has to:
- interrupt gpio : detect door open/close with interlock system. Select state.
    - with door open:
        - turn off heater, do not turn on
        - detect alignment key switch:
            - ON: blink internal red LED, allow laser interlock, turn on external green LED
            - OFF: blink external yellow LED, turn on internal red LED (if heater has been on), turn off external green LED
        - control laser heater normally
    - with door closed:
        - allow laser interlock, turn on external green LED, turn off yellow and red LEDs
        - control heaters normally
        - if heater is on, turn on external yellow
- interrupt serial : receive temperature setting from software (temperature and tolerance)
- timer 1 period: 
    - measure temperature in sampled volume, inside enclosure and inside laser enclosure
    - inform software about temperatures  
    - Heaters control: if T below tolerance: turn on until the temperature is >= tolerance
- timer 2 period: toggle LEDs.

## GPIO

- LEDs High state, No pull, open drain. LEDs to 12V with 620 Ohm.
- Pull up in AlignmentKey (PB8)
- EXTI in PC13 and PB9, fall and rise.

## Timer

TIM6 enabled with IRQ. Prescaler : 999. Counter PEriod 2999. Clock timers 12MHz.

## Connectivity

- I2C1 : I2C
- USART2 : IRQ, 115200 B/s

## NVIC

Priority : use 1-3 to allow HAL_Delay to work.

## CryLas VGA

- Interlock pin : (VGA-1) to Relay1 (GND on the other side). Also connects to GND of external red LED.
- Trigger mode in : (VGA-2) output TTL falling edge to swtich between external or internal trigger. PC-10, CN7-1, GPIO-Output, open-drain, high.
- Laser on : (VGA-3) output TTL falling edge to start/stop emission. PC-11, CN7-2, GPIO-Output, open-drain, high.
- Trigger signal : (VGA-6) to coax trigger signal.
- Laser ready : (VGA-9) input TTL: high means laser is ready. PC-12, CN7-3.
- Trigger mode out : (VGA-10) input TTL: low-external, high-internal. PD-2 CN7-4.
- Laser emission : (VGA-11) input TTL: high-emission. PA-15, CN7-17.
- GND : VGA-15.
