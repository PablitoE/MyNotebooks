# HoloICE

## Aims

- Quantify cloud microphysical processes using a high-speed holographic imaging system. 
- Apply to numerical models to assess the impact on modelled cloud properties.

### Motivation

Get insight on effects of cloud microphysics on precipitation formation, storm electrification processes, and radiative transfer for getting more realistic simulations of clouds. Secondary ice processes increase the number of ice particles in clouds by fragmenting existing ice particles in various ways:
- freezing and shattering of water drops,
- rime splintering,
- following ice crystal collisions,
- fragmentation of ice during sublimation.

### Objectives

1. Develop a state-of-the science high-speed holographic imaging system for laboratory studies of cloud microphysical processes
2. Develop a quantitative description of the rime splintering process
3. Develop a quantitative description of the droplet freezing-shattering process
4. Develop numerical representations of rime-splintering and droplet-freezing processes in a variety of frameworks, including the Met Office CASIM scheme.
5. Assess the impact of rime-splintering and droplet freezing-shattering processes on real cloud systems.
